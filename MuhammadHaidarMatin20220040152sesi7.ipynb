```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#memuat picture
image =cv2.imread('lionscale.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
#
plt.figure(figsize=(6,4))
plt.imshow(image,cmap='gray')
plt.title("Image original")
plt.axis('off')
plt.show()
```

```python
#image
plt.figure(figsize=(6,4))
plt.hist(image.ravel(),bins=256,range={0,256})
plt.title("Histogram")
plt.show()
```

```python
#no 2. Lakukan histogram equalitzon pada citra
equ=cv2.equalizeHist(image)
```

```python
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.show()
```

```python
#no 3. Hitung dan tampilkan histogram dari citra hasil equlization
plt.subplot(1,2,1)
plt.imshow(equ,cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Pastikan image sudah didefinisikan. Misalnya:
# image = cv2.imread('path/to/image.jpg', cv2.IMREAD_GRAYSCALE)

# Citra hasil equalization
equalized_image = cv2.equalizeHist(image)

# Menampilkan perbandingan citra asli dan citra hasil equalization
plt.figure(figsize=(12, 6))

# Menampilkan citra asli
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis('off')

# Menampilkan citra hasil equalization
plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Gambar Hasil Equalization")
plt.axis('off')

plt.show()
```
